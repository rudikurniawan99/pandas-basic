```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('dataset/Kickstarter+Projects/kickstarter_projects.csv')
df
```

```python
# pandas  data view
df.columns
df.index

# data with certain condition
# art project
paint_project = df[(df['Category'] == 'Art') & (df['Subcategory'] == 'Painting')]
paint_project.head(7)
```

```python
# get list of the Country
country_list = paint_project['Country'].unique()
country_list.tolist()
```

```python
# start new thread
tech_field = df[df['Category'] == 'Technology']
country_distribution = pd.DataFrame(tech_field['Country'].value_counts().reset_index())
country_distribution['percentage'] = (country_distribution['count']/country_distribution['count'].sum())*100
country_distribution.head(5)

```

```python
# Data Visualization
sns.catplot(
  data=country_distribution.head(5),
  x='Country',
  y='percentage',
  kind='bar'
)
```

```python
df['Deadline'] = pd.to_datetime(df['Deadline'])
df['Launched'] = pd.to_datetime(df['Launched'])
df.info()
```

```python
import datetime

df['Deadline'].dt.year.value_counts()
year_distribution = pd.DataFrame(data=df['Deadline'].dt.year.value_counts().reset_index())
year_distribution['percent'] = (year_distribution['count']/year_distribution['count'].sum())
year_distribution

```

```python
sns.scatterplot(
  data=year_distribution,
  x='Deadline',
  y='count',
  size='count'
)
```
